```python
def process_yearly_data(df):

    df_train = df.copy()
    
    df_train = df_train.loc[:, ~df_train.columns.str.contains('^Unnamed')]

    df_train['GL_TRANSACTION_EFFECTIVE_DT'] = pd.to_datetime(df_train['GL_TRANSACTION_EFFECTIVE_DT'])

    df_train['DAILY_CUMULATIVE'] = df_train.groupby('GL_TRANSACTION_EFFECTIVE_DT')['TRANSACTION_AMOUNT'].cumsum()

    monthly_totals = df_train.groupby(['MONTH', 'YEAR'])['TRANSACTION_AMOUNT'].sum().reset_index()
    monthly_totals.rename(columns={'TRANSACTION_AMOUNT': 'MONTHLY_TOTAL'}, inplace=True)

    df_merged = pd.merge(df_train, monthly_totals, on=['MONTH', 'YEAR'], how='left')

    df_merged['DAY'] = df_merged['GL_TRANSACTION_EFFECTIVE_DT'].dt.day

    df_pivot = df_merged.pivot_table(index='MONTH', columns='DAY', values='TRANSACTION_AMOUNT', aggfunc='sum')

    df_pivot['MONTHLY_TOTAL'] = monthly_totals.set_index('MONTH')['MONTHLY_TOTAL']
    df_pivot.columns = [f'Day {col}' if isinstance(col, int) else col for col in df_pivot.columns]

    return df_merged, df_pivot
```

```python
df_merged_total, df_pivot_total = process_yearly_data(df_total)
```

```python
def process_cumulative_data(df_pivot, df_merged, transaction_types, num_days):
    
    # Create Day columns dynamically based on num_days
    days_columns = [f'Day {i+1}' for i in range(num_days)]
    df_days = df_pivot[days_columns]
    
    # Filter and calculate cumulative sums for each transaction type within the first N days
    first_days_data = df_merged[df_merged['DAY'] <= num_days]
    results = []
    
    for t_type in transaction_types:
        type_data = first_days_data[first_days_data['FIN_SOURCE_TYPE_DESC'] == t_type]
        group_sum = type_data.groupby(['MONTH'])['TRANSACTION_AMOUNT'].sum().reset_index()
        group_sum.rename(columns={'TRANSACTION_AMOUNT': f'first_{num_days}_days_{t_type.replace(" ", "_")}_cost'}, inplace=True)
        results.append(group_sum)
    
    # Combine all transaction types into one dataframe
    final_results_df = pd.concat(results, axis=1)
    final_results_df = final_results_df.loc[:,~final_results_df.columns.duplicated()]
    
    # Merge cumulative days cost and transaction type costs
    result_df = pd.merge(df_days, final_results_df, on='MONTH', how='left')
    result_df = result_df.fillna(0)
    
    # Return the desired columns
    transaction_columns = [f'first_{num_days}_days_{t_type.replace(" ", "_")}_cost' for t_type in transaction_types]
    final_columns = days_columns + transaction_columns
    X = result_df[final_columns]
    
    return X

# Example usage:
transaction_types = ['INTERDISTRICT BILLING', 'WARRANTY CLAIMS', 'REPAIR ORDERS', 
                     'WALKER ACCOUNTS PAYABLE', 'ELECTRONIC INVOICES']
```

```python
results = []

predictions_dict = {}

for num_days in range(5, 22):  # From N=5 to N=21

    # Prepare X
    X = process_cumulative_data(df_pivot_total, df_merged_total, transaction_types, num_days)
    
    month_index = df_pivot_total.index
    X.set_index(month_index, inplace=True)
    
    data = X.copy()
    data['MONTHLY_TOTAL'] = df_pivot_total['MONTHLY_TOTAL']#.fillna(0)
        
    Y = data['MONTHLY_TOTAL']
    X = data.drop(columns=['MONTHLY_TOTAL'])
    
    train = data[data.index < '2024-01-01']
    test = data[data.index >= '2024-01-01') & (data.index < '2024-01-01')]

    X_train, Y_train = train.drop(columns=['MONTHLY_TOTAL']), train['MONTHLY_TOTAL']
    X_test, Y_test = test.drop(columns=['MONTHLY_TOTAL']), test['MONTHLY_TOTAL']

    # Train and evaluate the model
    model = LinearRegression()
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    
    # Store predictions and actuals for plotting
    predictions_dict[num_days] = (test.index, Y_pred, Y_test)

    # Calculate R² and RMSE
    r2 = r2_score(Y_test, Y_pred)
    rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
    
    results.append((num_days, r2, rmse))
```

```python
# Print results
for result in results:
    print(f"Days: {result[0]}, R2: {result[1]:.4f}, RMSE: {result[2]:.2f}")
```

```python
for num_days in predictions_dict:
    dates, y_pred, y_actual = predictions_dict[num_days]
    plt.figure(figsize=(10, 5))
    plt.plot(dates, y_actual, label='Actual', marker='o')
    plt.plot(dates, y_pred, label='Predicted', linestyle='--', marker='x')
    plt.title(f'Actual vs Predicted for N = {num_days} Days')
    plt.xlabel('Month')
    plt.ylabel('Monthly Total')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)  # Rotate dates for better visibility
    plt.tight_layout()
    plt.show()
```
