```python
def process_yearly_data(df, year):
    df_train = df.copy()

    df_train['GL_TRANSACTION_EFFECTIVE_DT'] = pd.to_datetime(df_train['GL_TRANSACTION_EFFECTIVE_DT'])

    df_train['DAILY_CUMULATIVE'] = df_train.groupby('GL_TRANSACTION_EFFECTIVE_DT')['TRANSACTION_AMOUNT'].cumsum()

    monthly_totals = df_train.groupby(['MONTH', 'YEAR'])['TRANSACTION_AMOUNT'].sum().reset_index()
    monthly_totals.rename(columns={'TRANSACTION_AMOUNT': 'MONTHLY_TOTAL'}, inplace=True)

    df_merged = pd.merge(df_train, monthly_totals, on=['MONTH', 'YEAR'], how='left')

    df_merged['DAY'] = df_merged['GL_TRANSACTION_EFFECTIVE_DT'].dt.day

    df_pivot = df_merged.pivot_table(index='MONTH', columns='DAY', values='TRANSACTION_AMOUNT', aggfunc='sum')

    df_pivot['MONTHLY_TOTAL'] = monthly_totals.set_index('MONTH')['MONTHLY_TOTAL']
    df_pivot.columns = [f'Day {col}' if isinstance(col, int) else col for col in df_pivot.columns]

    return df_merged, df_pivot
```

```python
df_pivot_filled = df_pivot.fillna(0)
```

```python
# Lists to store R² scores and RMSE values
r2_scores = []
rmse_values = []
day_range = range(5, 22)  # Days from 5 to 21

# Loop to iterate over 5 days, 6 days, up to 21 days
for days in day_range:
    day_columns = [f'Day {i}' for i in range(1, days + 1)]
    
    # Prepare the cleaned dataframe by replacing NaN with 0
    df_clean = df_pivot_filled[day_columns + ['MONTHLY_TOTAL']]
    
    # Separate features (X) and target (Y)
    X = df_clean[day_columns]
    Y = df_clean['MONTHLY_TOTAL']
    
    # Fit the linear regression model
    model = LinearRegression()
    model.fit(X, Y)
    
    # Predict the Monthly Cost
    Y_pred = model.predict(X)
    
    # Calculate R² score and RMSE
    r2_score = model.score(X, Y)
    rmse = np.sqrt(mean_squared_error(Y, Y_pred))
    
    # Store R² score and RMSE
    r2_scores.append(r2_score)
    rmse_values.append(rmse)
```

### GLM

```python
for days in range(5, 11):
    day_columns = [f'Day {i}' for i in range(1, days + 1)]
    
    # Prepare the cleaned dataframe by replacing NaN with 0
    df_clean = df_pivot_filled[day_columns + ['Monthly Cost']]
    
    # Separate features (X) and target (Y)
    X = df_clean[day_columns]
    Y = df_clean['Monthly Cost']
    
    # Fit the GLM model (Poisson as example)
    X = sm.add_constant(X)  # Add constant term for the intercept
    glm_model = sm.GLM(Y, X, family=sm.families.Gaussian())  # You can replace with other families if needed
    glm_results = glm_model.fit()
    
    # Predict the Monthly Cost
    Y_pred = glm_results.predict(X)
    
    # Plot Actual vs Predicted Monthly Costs for each model
    plt.scatter(df_clean.index, Y, color='red', marker='x', alpha=0.6, label='Actual' if days == 5 else "")
    plt.scatter(df_clean.index, Y_pred, alpha=0.6, label=f'Predicted ({days} days)')
    
    # Add a line to better visualize the comparison
    plt.plot(df_clean.index, Y_pred, linestyle='--', alpha=0.6)
```
